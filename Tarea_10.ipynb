```
class Queue:
    def __init__( self ):
        self.__data = list()

    def is_empty( self ):
        return ( len(self.__data) == 0 )

    def lenght( self ):
        return len(self.__data)

    def enqueue( self, elem ):
        self.__data.append(elem)

    def dequeue( self ):
        if self.is_empty():
            print("La cola está vacía")
        else:
            return self.__data.pop(0)

    def to_string( self ):
        cadena = ""
        for elem in self.__data:
            cadena = cadena + " | " + str(elem)
        cadena = cadena + " |"
        print(cadena)

class BoundedPriorityQueue:
    def __init__( self, niveles ):
        self.__data = [ Queue() for x in range(niveles) ]
        self.__size = 0

    def is_empty( self ):
        return self.__size == 0

    def lenght( self ):
        return self.__size

    def enqueue(self, prioridad, elem):
        if prioridad < len(self.__data) and prioridad >=0:
            self.__data[ prioridad ].enqueue( elem )
            self.__size += 1

    def dequeue( self ):
        if not self.is_empty():
            for nivel in self.__data:
                if not nivel.is_empty():
                    return nivel.dequeue()
                    self.__size -= 1

    def to_string( self ):
        print("Cola:")
        i = 0
        for nivel in self.__data:
            print(f"Nivel {i}", end="")
            nivel.to_string()
            i += 1
            
maestres = {"Prioridad" : 4,
            "Descripcion" : "Maestre",
            "Personas" : ("Juan P.", "Diego H.")}

niños = {"Prioridad" : 2,
        "Descripcion" : "Niños",
        "Personas" : ("Santi H.", "Angel H.")}

mecanicos = {"Prioridad" : 4,
            "Descripcion" : "Mecanicos",
            "Personas" : ("Diana T.", "Maria Z.")}

niñas = {"Prioridad" : 1,
            "Descripcion" : "Niñas",
            "Personas" : ("Valeria T.", "Dana F.")}

ancianos = {"Prioridad" : 2,
            "Descripcion" : "3ra edad",
            "Personas" : ("Martín A.", "Karla E.")}

hombres = {"Prioridad" : 3,
            "Descripcion" : "Hombres",
            "Personas" : ("Pepe D.", "Fernando H.")}

mujeres = {"Prioridad" : 3,
            "Descripcion" : "Mujeres",
            "Personas" : ("Diana E.", "Sucy M.")}

vigia = {"Prioridad" : 4,
            "Descripcion" : "Vigía",
            "Personas" : ("Diana T.", "Maria Z.")}

timonel = {"Prioridad" : 4,
            "Descripcion" : "Timonel",
            "Personas" : ("Eduardo T.")}

capitan = {"Prioridad" : 5,
            "Descripcion" : "Capitan",
            "Personas" : ("Minamitsu Murasa")}

cpa = BoundedPriorityQueue( 7 )
cpa.enqueue( maestres["Prioridad"], maestres )
cpa.enqueue( niños["Prioridad"], niños )
cpa.enqueue( mecanicos["Prioridad"], mecanicos )
cpa.enqueue( niñas["Prioridad"], niñas)
cpa.enqueue( ancianos["Prioridad"], ancianos)
cpa.enqueue( hombres["Prioridad"], hombres)
cpa.enqueue( mujeres["Prioridad"], mujeres)
cpa.enqueue( vigia["Prioridad"], vigia)
cpa.enqueue( timonel["Prioridad"], vigia)
cpa.enqueue( capitan["Prioridad"], capitan)
cpa.to_string()

#Evacuación del barco
print("-----------------------------------")
print("Iniciando evacuación del barco...\n")

for x in range(10):
    print("Evacuando: ", end="")
    print(cpa.dequeue())
    
print("\nEl barco ha sido completamente evacuado...")
```
