```
class PriorityQueue:
    def __init__( self ):
        self.__data = list()
    
    def is_empty( self ):
        return len(self.__data) == 0
    
    def lenght( self ):
        return len(self.__data)
    
    def enqueue( self, prioridad, elem):
        if self.is_empty():
            self.__data.append({'prioridad':prioridad, 'Elemento':elem})
            
        else:
            i = 0
            while i < len(self.__data):
                if self.__data[i]['prioridad'] > prioridad:
                    break
                i += 1
            
            if i < len(self.__data):
                self.__data.insert(i, {'prioridad':prioridad, 'Elemento':elem})
            
            elif i >= len(self.__data):
                self.__data.append({'prioridad':prioridad, 'Elemento':elem})
                
    
    def dequeue(self):
        if not self.is_empty():
            return self.__data.pop(0)
        
        else:
            print("La cola está vacía")
            
    def to_string( self ):
        if self.is_empty():
            print("La cola está vacía")
        
        else:
            for elem in self.__data:
                print(elem)
                
                
pq = PriorityQueue()

print("-----Funcion enqueue-----")
pq.enqueue(3, 'Mario')
pq.enqueue(1, "Pepe")
pq.enqueue(5, "Donald")
pq.enqueue(1, "Diana")
pq.enqueue(2, "Abril")
pq.enqueue(6, "Fer")
pq.enqueue(8, "Gerardo")
pq.enqueue(7, "Giselle")
pq.enqueue(0, "Bart")
pq.to_string()

print("\n-----Funcion lenght-----")
print("Tamaño de la lista:", pq.lenght())

print("\n-----Funcion dequeue-----")
print(pq.dequeue())
print("\nNueva Cola:")
pq.to_string()
print("Tamaño de la lista:", pq.lenght())
```
