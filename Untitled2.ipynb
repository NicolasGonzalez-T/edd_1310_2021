```
class NodoDoble:
    def __init__( self, value, siguiente = None, anterior = None ):
        self.data = value
        self.next = siguiente
        self.prev = anterior
        
class DoubleLinkedList:
    def __init__( self ):
        self.__head = NodoDoble( None, None, None )
        self.__tail = NodoDoble( None, None, None )
        self.__head.next = self.__tail
        self.__tail.prev = self.__head
        self.__size = 0
        
    def get_size( self ):
        curr_node = self.__head
        while curr_node != None:
            self.__size += 1
            curr_node = curr_node.next
        return self.__size
    
    def is_empty( self ):
        return self.__head.data == None
     
    def append( self, value):
        new_node = NodoDoble( value, None, None )

        if self.is_empty():
            new_node.next = None
            new_node.prev = None
            self.__head = new_node
            self.__size += 1
            
        else:
            curr_node = self.__head
            while curr_node.next != None:
                curr_node = curr_node.next
                
            new_node.prev = curr_node
            curr_node.next = new_node
            self.__tail = curr_node.next
            self.__size += 1
            
            
            
    def find_from_head( self, value ):
        curr_node = self.__head
        posicion = 0
        while curr_node.data != value and curr_node.next != None:
            posicion += 1
            curr_node = curr_node.next
            
        if curr_node.data == value:            
            pass
        
        else:
            posicion = "No se encontró ninguna coincidencia"
        
        return posicion
    
    def find_from_tail( self, value):
        curr_node = self.__tail
        posicion = 0
        while curr_node.data != value and curr_node.prev != None:
            posicion += 1
            curr_node = curr_node.prev
            
        if curr_node.data == value:
            pass
        
        else:
            posicion = "No se encontró ninguna coincidencia"
            
        return posicion
    
    def remove_from_head( self, value):
        curr_node = self.__head
        if self.__head.data == value:
            self.__head = self.__head.next
        
        else:
            prev_node = None
            while curr_node.data != value and curr_node.next != None:
                prev_node = curr_node
                curr_node = curr_node.next
            
            if curr_node.data == value:
                curr_node.prev = prev_node
                prev_node.next = curr_node.next
            
            else:
                print("No se encontró ninguna coincidencia")
    
    def remove_from_tail( self, value ):
        curr_node = self.__tail
        if self.__tail.data == value:
            self.__tail = self.__tail.prev
            
        else:
            prev_node = None
            while curr_node.data != value and curr_node.prev != None:
                curr_node = curr_node.prev
                prev_node = curr_node.prev
                
            if curr_node.data == value:
                curr_node.prev = prev_node
                prev_node.next = curr_node.next
            
            else:
                print("No se encontró ninguna coincidencia")
            
    def transversal( self ):
        curr_node = self.__head
        while curr_node != None:
            print(f"{curr_node.data} ---> ", end="")
            curr_node = curr_node.next
            
        print("")
        
    def reverse_transversal( self ):
        curr_node = self.__tail
        while curr_node != None:
            print(f"{curr_node.data} ---> ", end="")
            curr_node = curr_node.prev
            
        print("")
            
    
            
        
        
dll = DoubleLinkedList()
dll.append(5)
dll.append(10)
dll.append(15)
dll.append(20)
dll.append(10)
dll.append(15)
dll.append(30)

#Método find_from_head
print(dll.find_from_head(10))

#Método find_from_tail
print(dll.find_from_tail(10))

#Método transversal
dll.transversal()

#Método remove from head
dll.remove_from_head(10)
dll.transversal()

#Método remove from tail
dll.remove_from_tail(15)
dll.transversal()


dll.reverse_transversal()
```
